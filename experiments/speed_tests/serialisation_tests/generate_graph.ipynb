```python
import numpy as np

preparations = []
serialisations = []

for i in range(30):
    results = !./test.sh

    result_pairs = [tuple([int(y) for y in x.split(" ")]) for x in results]
    [preparation, serialisation] = list(zip(*result_pairs))
    preparation = list(preparation)
    serialisation = list(serialisation)
    preparations.append(preparation)
    serialisations.append(serialisation)
    print(preparation)
    print(serialisation)

preparations = np.array(preparations)
serialisations = np.array(serialisations)

avg_preparation = np.mean(preparations, axis=0)
avg_serialisation = np.mean(serialisations, axis=0)

print(avg_preparation)
print(avg_serialisation)
```

```python
import numpy as np

preparations = []
serialisations = []

for i in range(30):
    results = !./test_str.sh

    result_pairs = [tuple([int(y) for y in x.split(" ")]) for x in results]
    [preparation, serialisation] = list(zip(*result_pairs))
    preparation = list(preparation)
    serialisation = list(serialisation)
    preparations.append(preparation)
    serialisations.append(serialisation)
    print(preparation)
    print(serialisation)

preparations = np.array(preparations)
serialisations = np.array(serialisations)

avg_preparation_str = np.mean(preparations, axis=0)
avg_serialisation_str = np.mean(serialisations, axis=0)

print(avg_preparation_str)
print(avg_serialisation_str)
```

```python
from matplotlib import pyplot as plt

fig, (ax1, ax2) = plt.subplots(2, 1)

x_axis = list([1, *range(5, 101, 5)])
ax1.plot(x_axis, avg_preparation, color="C0", label="argument evaluation")
ax1.plot(x_axis, avg_serialisation, color="C1", label="argument serialisation")
ax2.plot(x_axis, avg_preparation_str, color="C0", label="argument evaluation")
ax2.plot(x_axis, avg_serialisation_str, color="C1", label="argument serialisation")

ax1.set_xlabel("number of function parameters")
ax2.set_xlabel("number of function parameters")
ax1.set_ylabel("time ($\mu$s)")
ax2.set_ylabel("time ($\mu$s)")
ax1.set_title("Functions with integer parameters")
ax2.set_title("Functions with string parameters")
ax1.legend()
ax2.legend()
plt.tight_layout()

plt.savefig("serialisation.pdf")
```
